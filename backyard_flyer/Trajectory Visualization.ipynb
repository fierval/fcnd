## Visualize Drone Trajectory ##
1. Parse the log file
2. Extract local position messages
3. Plot progression

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
log_file = r'C:\Git\udacity\backyard_flyer\Logs\TLog.txt'
```

### Parse the Log & Extract Local Postion Progression ###

```python
df_log = pd.read_csv(log_file, header=None, usecols=[0,2,3,4])
df_log = df_log[df_log[0] == 'MsgID.LOCAL_POSITION']
df_log.drop(0, axis=1, inplace=True)
df_log.columns = ['North', 'East', 'Height']
df_log['Height'] = -df_log['Height'] 
df_log.reset_index(drop=True, inplace=True)
df_log = df_log.applymap(float) 
```

```python
df_log.head()
```

### Plot it ###

```python
plt.rcParams['figure.figsize'] = (20, 20)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(df_log.North, df_log.East, df_log.Height)
ax.set_xlabel('North')
ax.set_ylabel('East')
_ = ax.set_zlabel('Height')
```
